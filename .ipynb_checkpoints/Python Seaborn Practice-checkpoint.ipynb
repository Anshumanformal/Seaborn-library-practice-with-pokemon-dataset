# Importing libraries.

```python
import pandas as pd
# Matplotlib for additional customization
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Seaborn for plotting and styling
import seaborn as sns
```

```python
# Read dataset
df = pd.read_csv('Pokemon.csv', encoding = 'latin')
df
```

```python
# Display first 5 observations
df.head()
```

# Seaborn's plotting functions 

# Different plots used in this notebook:-
### 1. lmplot()
### 2. boxplot()
### 3. violinplot()
### 4. swarmplot()
### 5. Heatmaps - for matrix like data.
### 6. Histograms
### 7. Bar plots
### 8. Factor plots(or catplots)
### 9. Density plots
### 10. Joint distribution plots

```python
sns.lmplot(x ="Attack", y = "Defense", data = df)
```

```python
sns.lmplot(x='Attack', y='Defense', data=df,
           fit_reg=False, # No regression line
           hue='Stage')   # Color by evolution stage
```

# Boxplot 

```python
sns.boxplot(data = df)
```

```python
# Pre-format DataFrame
stats_df = df.drop(columns = ['Total', 'Stage', 'Legendary'])

#Another syntax.
#stats_df = df.drop(['Total', 'Stage', 'Legendary'], axis = 1)
 
# New boxplot using stats_df
sns.boxplot(data=stats_df)
```

```python
plt.figure(figsize = (10,10))

# Set theme
sns.set_style('whitegrid')
 
# Violin plot
sns.violinplot(x='Type 1', y='Attack', data=df)
```

# Using seaborn custom colour pallettes by utilizing ordered python lists. 

```python
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]
```

```python
# Violin plot with Pokemon color palette
plt.figure(figsize = (18,10))
sns.violinplot(x='Type 1', y='Attack', data=df, 
               palette=pkmn_type_colors) # Set color palette
```

# Swarm Plots

```python
# Swarm plot with Pokemon color palette
plt.figure(figsize = (18,10))
sns.swarmplot(x='Type 1', y='Attack', data=df, 
              palette=pkmn_type_colors)
```

# Overlaying swarm and violin plots. 

```python
# Set figure size with matplotlib
plt.figure(figsize=(10,6))
 
# Create plot
sns.violinplot(x='Type 1',
               y='Attack', 
               data=df, 
               inner=None, # Remove the bars inside the violins
               palette=pkmn_type_colors)
 
sns.swarmplot(x='Type 1', 
              y='Attack', 
              data=df, 
              color='k', # Make points black
              alpha=0.7) # and slightly transparent
 
# Set title with matplotlib
plt.title('Attack by Type')
```

### This chart tells us how Attack values are distributed across different PokÃ©mon types. 

# Melting the different attack columns into one column. 

```python
# Melt DataFrame
melted_df = pd.melt(stats_df, 
                    id_vars=["Name", "Type 1", "Type 2"], # Variables to keep
                    var_name="Stat") # Name of melted variable
melted_df.head()
```

# Heatmaps - used to visualize matrix like data. 

```python
# Calculate correlations
corr = stats_df.corr()

plt.figure(figsize = (18,10))
    
# Heatmap
sns.heatmap(corr)
```

# Histograms - used to plot the distributions of numeric variables. 

```python
# Distribution Plot (a.k.a. Histogram)
plt.figure(figsize = (18,10))
sns.distplot(df.Attack)
```

# Bar plots - used to visualize the distributions of categorical variables. 

```python
plt.figure(figsize = (18,10))

# Count Plot (a.k.a. Bar Plot)
sns.countplot(x='Type 1', data=df, palette=pkmn_type_colors)

# Rotate x-labels
plt.xticks(rotation=-45)
```

# Factor plots(now called catplots) - used to separate plots by categorical classes.

```python
# Factor Plot
g = sns.catplot(x='Type 1', 
                   y='Attack', 
                   data=df, 
                   hue='Stage',  # Color by stage
                   col='Stage',  # Separate by stage
                   kind='swarm') # Swarmplot
 
# Rotate x-axis labels
g.set_xticklabels(rotation=-45)
 
# Doesn't work because only rotates last plot
# plt.xticks(rotation=-45)    
```

# Density plots - used to display the distribution between two variables. 

```python
plt.figure(figsize = (18,10))
# Density Plot
sns.kdeplot(df.Attack, df.Defense)
```

# Joint distribution plot - used to combine information from scatter plots and histograms to give you detailed information for bi-variate distributions. 

```python
# Joint Distribution Plot
sns.jointplot(x='Attack', y='Defense', data=df)
```
